# Central limit theorem by Yourself

I've chosen the **beta distribution**

Import block:

```python
import numpy as np
import pandas as pd
from scipy.stats import beta,norm
import matplotlib.pylab as plt
from __future__ import division

%matplotlib inline
plt.style.use('ggplot')
```

# Part I
Generating sample.

```python
a = 0.4
b = 0.5

betas = beta(a=a,b=b)
sample = betas.rvs(1000)
```

# Graphing
Building a histogram and the theoretical distribution density:

```python
x = np.linspace(0,1,100) #limits

pdf = betas.pdf(x)
plt.plot(x,pdf)

plt.hist(sample, bins=40, normed=True) #amount is 40
plt.ylabel('$PDF(x)$') #labeling
plt.xlabel('$x$')
```

# Part II
Generating 1000 values of sample means for several values of n (5, 10, 50):



```python
#Sample of 5:
sample_5 = np.array([np.mean(np.random.choice(sample,5)) for i in xrange(1000)])

#Sample of 10:
sample_10 = np.array([np.mean(np.random.choice(sample,10)) for i in xrange(1000)])    

#Sample of 50:
sample_50 = np.array([np.mean(np.random.choice(sample,50)) for i in xrange(1000)])
```

# Central limit theorem

```python
mu = a/(a+b)
dispersion = (a*b)/((a+b)**2 * (a+b+1))
```

 $\frac{DX}{n}$ - dispersion.In the scipy the normal distribution is given by the mathematical expectation and the standard deviation (and not the variance), then as a parameter we will transfer $\sqrt{\frac{DX}{n}}$

**For distribution of sample means with sample size of 50:**

```python
norm50 = norm(loc=mu, scale=np.sqrt(dispersion/50))
pdf50 = norm50.pdf(x)
plt.hist(sample_50, bins=30, normed=True)
plt.plot(x,pdf50)
plt.title('$N = 50$')
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

**For distribution of sample means with sample size of 10:**

```python
norm10 = norm(loc=mu, scale=np.sqrt(dispersion/10))
pdf10 = norm10.pdf(x)
plt.hist(sample_10, bins=30, normed=True)
plt.plot(x,pdf10)
plt.title('$N = 10$')
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

**For distribution of sample means with sample size of 5:**

```python
norm5 = norm(loc=mu, scale=np.sqrt(dispersion/5))
pdf5 = norm5.pdf(x)
plt.hist(sample_5, bins=30, normed=True)
plt.plot(x,pdf5)
plt.title('$N = 5$')
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

# Conclusion

**Central limit theorem** is working! The accuracy increases with increasing the number of elements for constructing the sample mean(N).
