```python
import pandas as pd
import random
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm
import math
from scipy import special
```

РўРµРѕСЂРµС‚РёС‡РµСЃРєРёРµ РїР»РѕС‚РЅРѕСЃС‚Рё РІРµСЂРѕСЏС‚РЅРѕСЃС‚Рё

```python
# Р¤СѓРЅРєС†РёСЏ РїР»РѕС‚РЅРѕСЃС‚Рё РІРµСЂРѕСЏС‚РЅРѕСЃС‚Рё РҐРё-РєРІР°РґСЂР°С‚
def f (x):                  
    return ((0.5)**(k/2))*x**(k/2-1)*math.exp(-x/2)/special.gamma(k/2)

# Р¤СѓРЅРєС†РёСЏ РїР»РѕС‚РЅРѕСЃС‚Рё РІРµСЂРѕСЏС‚РЅРѕСЃС‚Рё РЅРѕСЂРјР°Р»СЊРЅРѕРіРѕ СЂР°СЃРїСЂРµРґРµР»РµРЅРёСЏ РґР»СЏ Р¦РџРў (РґРёСЃРїРµСЂСЃРёСЏ РґРµР»РµРЅРЅР°СЏ РЅР° N)
# Р”Р»СЏ РҐРё-РєРІР°РґСЂР°С‚ СЂР°СЃРїСЂРµРґРµР»РµРЅРёСЏ РјР°С‚РѕР¶РёРґР°РЅРёРµ СЂР°РІРЅРѕ k, Р° РґРёРїРµСЂСЃРёСЏ 2 k
#Р­С‚РѕС‚ РєСѓСЃРѕРє РєРѕРґР° Р±СѓРґСѓ РґСѓР±Р»РёСЂРѕРІР°С‚СЊ РєР°Р¶РґС‹Р№ СЂР°Р· РІС‹Р·С‹РІР°СЏ РґР°РЅРЅСѓСЋ С„СѓРЅРєРёСЋ, С‚Р°Рє РєР°Рє РїР°СЂР°РјРµС‚СЂС‹ N Рё sigma Р±СѓРґСѓС‚ РјРµРЅСЏС‚СЊСЃСЏ

N = 5
k = 5.0  # СЃС‚РµРїРµРЅРµР№ СЃРІРѕР±РѕРґС‹  СЂР°СЃРїСЂРµРґРµР»РµРЅРёСЏ РҐРё-РєРІР°РґСЂР°С‚
sigma = math.sqrt(2*k/N) 
mu = k
def g (x):
    return math.exp(-0.5*((x-mu)/sigma)**2) / (sigma * math.sqrt(2*math.pi))

#print('f=',f(2))
#print('g=',g(2))
```

Р‘СѓРґРµРј СЂР°СЃСЃРјР°С‚СЂРёРІР°С‚СЊ СЂР°СЃРїСЂРµРґРµР»РµРЅРёРµ РҐРё-РєРІР°РґСЂР°С‚ СЃ 5 СЃС‚РµРїРµРЅСЏРјРё СЃРІРѕР±РѕРґС‹. 

```python
vib = 1000
viborka = np.random.chisquare(k,vib)
#print (viborka)
#type (viborka)

bins = round(10+math.sqrt(len(viborka))) # Ñ„Ð¾Ñ€Ð¼ÑƒÐ»Ð¾Ð¹ Ð¡Ñ‚ÑÐ´Ð¶ÐµÑÑÐ° Ð´Ð»Ñ Ð³Ñ€ÑƒÐ¿Ð¸Ñ€Ð¾Ð²ÐºÐ¸ Ð´Ð°Ð½Ð½Ñ‹Ñ…
#print ( bins_number )
n, bins, patches = plt.hist(viborka, bins, normed=1, facecolor='blue', alpha=0.4) 

# Ð”Ð»Ñ Ð¿Ð¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ñ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ° Ñ„ÑƒÐ½-Ñ†Ð¸Ð¸. Ð˜Ð½Ñ‚ÐµÑ€Ð²Ð°Ð» Ð¸Ð·Ð¼ÐµÐ½ÐµÐ½Ð¸Ñ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ð¾Ð¹ Ð¿Ð¾ Ð¾ÑÐ¸ X
xmin = 0.0
xmax = 17.0
dx = 0.01   # Ð¨Ð°Ð³ Ð¼ÐµÐ¶Ð´Ñƒ Ñ‚Ð¾Ñ‡ÐºÐ°Ð¼Ð¸

xlist = mlab.frange (xmin, xmax, dx) # ÑÐ¿Ð¸ÑÐ¾Ðº ÐºÐ¾Ð¾Ñ€Ð´Ð¸Ð½Ð°Ñ‚ Ð¿Ð¾ Ð¾ÑÐ¸ X Ð½Ð° Ð¾Ñ‚Ñ€ÐµÐ·ÐºÐµ [-xmin; xmax], Ð²ÐºÐ»ÑŽÑ‡Ð°Ñ ÐºÐ¾Ð½Ñ†Ñ‹
ylist = [f(x) for x in xlist] # Ð’Ñ‹Ñ‡Ð¸ÑÐ»Ð¸Ð¼ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ñ„ÑƒÐ½ÐºÑ†Ð¸Ð¸ Ð² Ð·Ð°Ð´Ð°Ð½Ð½Ñ‹Ñ… Ñ‚Ð¾Ñ‡ÐºÐ°Ñ…

plt.plot(xlist, ylist, 'r--')
plt.ylabel('Ð§Ð°ÑÑ‚Ð¾Ñ‚Ð° Ñ€Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ñ')
plt.title('Ð Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð¥Ð¸-ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚ n=1000')
plt.show()
```
