```python
import numpy as np
import math
from numpy import linalg
from matplotlib import pylab as plt
%matplotlib inline

f = lambda x: np.sin(x/5.0)*np.exp(x/10.0) + 5.0 * np.exp(-x/2.0)


A = np.array([[1, 1], [15, 1]])
b = np.array([f(1), f(15)])

x = np.linalg.solve(A,b)
print x
print (np.dot(A,x) == b)

listik = []
listik2 = []

for i in range(15):
    listik2.append(i)
    listik.append(f(i))
    
plt.plot(listik2, listik)
plt.plot(x)
```

```python
A = np.array([[1,1,1],[1,8,64],[1,15,225]])
b = np.array([f(1), f(8), f(15)])

x = np.linalg.solve(A,b)
print (np.dot(A,x) == b)

plt.plot(listik2, listik)
plt.plt
```
